```python
import scvelo as scv
import anndata
import splicejac as sp
```

```python
adata = anndata.read("hsc_cellcommun.h5ad")
```

```python
adata.obs['clusters']=adata.obs['cell_type']
```

```python
scv.pp.filter_genes(adata, min_shared_counts=20)
scv.pp.normalize_per_cell(adata)
scv.pp.filter_genes_dispersion(adata, n_top_genes=2000)
scv.pp.log1p(adata)
```

```python
sp.tl.estimate_jacobian(adata, n_top_genes=20)
```

```python
sp.pl.visualize_jacobian(adata,
                         panel_height=3,
                         panel_length=3.7,
                         pan_per_row=4,
                         fontsize=10,
                         cmap='RdBu_r',
                         showfig=True,)
```
